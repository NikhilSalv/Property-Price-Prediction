```python
# The Eexploratory Data Analysis is completed.
# And we will perform random sampling on our model and build Linear regression on our datset.
```

```python
pp_x = pp.iloc[:,1:71]
pp_y = pp.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
pp_x_train, pp_x_test,pp_y_train,pp_y_test = train_test_split(pp_x,pp_y,test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(pp_x_train,pp_y_train)
```

```python
pred_test = reg.predict(pp_x_test)
```

```python
error = abs(pp_y_test - pred_test)
```

```python
import numpy as np
np.mean(error)
```

```python
MAPE = np.mean(np.abs(error * 100/ pp_y_test))
MAPE
```

```python
Accuracy = 100 - MAPE
Accuracy
```
